```python
## load in required modules:
import numpy as np
from numpy import genfromtxt
import pandas as pd
import csv
import matplotlib.pyplot as plt
from sklearn import preprocessing


def normalizeData(inDF):
    x = inDF[1:len(inDF)]
    min_max_scaler = preprocessing.MinMaxScaler()
    normalizedDF = pd.DataFrame(min_max_scaler.fit_transform(x))
    return normalizedDF.values

testCSV = genfromtxt('/Users/awuzaso/Downloads/test.csv', delimiter=',')
trainCSV = genfromtxt('/Users/awuzaso/Downloads/train.csv', delimiter=',')
```

```python
## read data into a Numpy array
test_df = csv.reader(open('/Users/awuzaso/Downloads/test.csv', 'rb'),delimiter=',')
train_df = csv.reader(open('/Users/awuzaso/Downloads/train.csv', 'rb'),delimiter=',')

# Read into headless dfs.
train_set = np.array(list(train_df))[1:,0:13].astype('float')
test_set = np.array(list(test_df))[1:,0:13].astype('float')

# Initializing train set
trainSet_burn = []
for i in train_set:
    if i[12] != 0:
        trainSet_burn.append(i)    
train_burn_array = np.array(trainSet_burn)

# Initializing testset
testSet_burn = []
for i in test_set:
    if i[12] != 0:
        testSet_burn.append(i)    
test_burn_array = np.array(testSet_burn)
print(len(train_burn_array))
print(len(test_burn_array))
print("///////")
# Initializing the bias term
nrow = train_burn_array.shape[0]
b0 = np.ones( (nrow,1) )


# DFs
# X Values
df_test_X = train_burn_array[:,0].reshape(-1, 1)
df_test_Y = train_burn_array[:,1].reshape(-1, 1)
df_test_month = train_burn_array[:,2].reshape(-1, 1)
df_test_day = train_burn_array[:,3].reshape(-1, 1)
df_test_FFMC = train_burn_array[:,4].reshape(-1, 1)
df_test_DMC = train_burn_array[:,5].reshape(-1, 1)
df_test_DC = train_burn_array[:,6].reshape(-1, 1)
df_test_ISI = train_burn_array[:,7].reshape(-1, 1)
df_test_temp = train_burn_array[:,8].reshape(-1, 1)
df_test_RH = train_burn_array[:,9].reshape(-1, 1)
df_test_wind = train_burn_array[:,10].reshape(-1, 1)
df_test_rain = train_burn_array[:,11].reshape(-1, 1)

# Get row number
train_row = df_test_X.shape
test_row = test_set.shape[0]

# X Value
X = np.concatenate((b0,df_test_X,df_test_Y,df_test_month,df_test_day,
                   df_test_FFMC,df_test_DMC,df_test_DC,df_test_ISI
                   ,df_test_temp,df_test_RH, df_test_wind,
                   df_test_rain),axis=1)
X_no_b0 = np.concatenate((df_test_X,df_test_Y,df_test_month,df_test_day,
                   df_test_FFMC,df_test_DMC,df_test_DC,df_test_ISI
                   ,df_test_temp,df_test_RH, df_test_wind,
                   df_test_rain),axis=1)
#X = np.concatenate((b0,df_test_temp,df_test_X,df_test_Y ),axis=1)
#X_L = np.concatenate((df_test_temp,df_test_X,df_test_Y ),axis=1)


# Y Value
df_test_burn = train_burn_array[:,12].T
Y = df_test_burn

# Calculate weights through the analytical solution:
w_list = np.dot(np.linalg.inv(np.dot(X.T,X)),np.dot(X.T,Y))
#print(w_list)
#print(w_list.shape)

print("Implementation weights: \n %s " %(w_list))
print("//////////////////")
z,resid,rank,sigma = np.linalg.lstsq(X,Y)
print("Numpy func weights: \n %s " %(z))
print("Numpy func resid: \n %s" %(resid))


# Generating the regression model:
wt = w_list[1:len(w_list)]

OLS_model = (wt.T).dot(X_no_b0.T) + w_list[0]

#print(X.shape)
#print(Y.shape)
#print(w_list.shape)

RSS = ((Y - X.dot(w_list)).T).dot(Y - X.dot(w_list))
print("Implementation RSS: %s" %(RSS))
'''
plt.plot(X_no_b0[:,0], Y, 'o', label='Original data', markersize=10)
plt.plot(X_no_b0[:,0], OLS_model, 'r', label='Fitted line')
plt.show()
'''
```
