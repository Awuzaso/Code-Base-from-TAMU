```python
# Import libraries.
import os
import math
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing

# Read in images.


# PCA
def pca(X,dropVal):
    
    # get matrix dimensions
    num_data, dim = X.shape
    # center data
    #mean_X = X.mean(axis=0)
    #X = X - mean_X
    M = np.dot(X, X.T) # covariance matrix
    e, U = np.linalg.eigh(M) # calculate eigenvalues and eigenvectors
    orderedE = e[::-1]
    orderedU = U[::-1]
    #print(orderedU)
    dimStart = dropVal-1
    orderedU_comp = orderedU#orderedU[dimStart:len(orderedU)]
    #print(orderedU_comp.shape)
    #print(orderedE)
    tmp = np.dot(X.T, orderedU_comp.T).T
    V = tmp[::-1] # reverse since the last eigenvectors are the ones we want
    S = np.sqrt(e)[::-1] #reverse since the last eigenvalues are in increasing order
            
    #return V,S,mean_X
    return V,S
```

```python
notebook_path = os.path.abspath("CSCE_633_HW_3_Q2")
realPath = '/Users/awuzaso/Downloads/yalefaces'
imagePath_NG = [os.path.join(realPath, f) for f in os.listdir(realPath)]
imageMatrix_NG = np.array([np.array(Image.open(im, 'r')).flatten()
                 for im in imagePath_NG], 'f')
print(imageMatrix_NG[0])
print(imageMatrix_NG[0].shape)
plt.gray
pca_NG,pca_NG_e = pca(imageMatrix_NG,14)

iterVal = 0
for i in range(10):
    
    plt.imshow(pca_NG[i].reshape(243,320),cmap='gray')
    plt.axis('off')
    
    
    

    fileName = os.path.join(os.path.dirname(notebook_path), str(iterVal))
    
    fileName = '/Users/awuzaso/Documents/codingProjects/python/CSCE_633_HWs/'+ str(iterVal) +'eigenFace.png'
    plt.savefig(fileName)
    Image.open(fileName).save(fileName,'PNG')
    iterVal = iterVal+1
```
