```python
import numpy as np

class rootObject:
    
    def __init__(self,bestSplitIndex,bestSplitVal,bestSplitGroups):
        self.bestSplitIndex = bestSplitIndex
        self.bestSplitVal = bestSplitVal
        self.bestSplitGroups = bestSplitGroups
        self.leftBranch = None
        self.rightBranch = None

class treeObject:
    
    def __init__(self, dataset, bestSplitVal,in_max_depth,in_size,in_depth,in_sc):
        #print("Initializing tree object")
        self.dataset = dataset
        self.sc = in_sc
        bI,bS,bG = self.getBestSplit(self.dataset,bestSplitVal,self.sc)
        self.rootNode = rootObject(bI,bS,bG)
        self.max_depth = in_max_depth
        self.min_size = in_size
        self.depth = in_depth
        
        '''
        self.max_depth = 2
        self.min_size = 2
        self.depth = 1
        '''
        self.split(self.rootNode,self.max_depth,self.min_size,self.depth,self.sc)

    # Select the best split point for a dataset
    def getBestSplit(self,dataset,bSplitVal,splitCrit):
        #print("Getting the Best Split")
        #1) Get set of output class values.
        indices = len(dataset[0])-1
        class_values = list(set(row[indices] for row in dataset))
        
    
        #2) Initialize best split values
        b_index = bSplitVal
        b_value = bSplitVal
        b_score = bSplitVal
        b_groups = None
    
        #3) Iterate through column.
        for index in range(indices):
        
            #3) Iterate through rows.
            for row in dataset:
            
                #3.1) Get splits.
                groups = self.getTestSplit(index, row[index], dataset)
                score = 0
                #3.2) Get split crit
                if(splitCrit == "Gini"):
                    score = self.getGiniIndex(groups, class_values)
                if(splitCrit == "Entropy"):
                    score = self.getEntropy(groups, class_values)
                
            #3.3)
                if score < b_score:
                    b_index, b_value, b_score, b_groups = index, row[index], score, groups
    
        return b_index, b_value, b_groups
    
    
    # Split a dataset based on an attribute and an attribute value
    def getTestSplit(self, index, value, dataset):
        #print("Getting Test Split")
        left_group = list()
        right_group = list()
        
        for row in dataset:
            if row[index] < value:
                left_group.append(row)
            else:
                right_group.append(row)
        return left_group, right_group
    
    
    
    
    def getGiniIndex(self,groups, classes):
        #print("Getting Gini Index")
        # 1) Get count of all instances in split groups.
        n_instances = 0
        for group in groups:
            n_instances += len(group)

        # Old:
        #n_instances = float(sum([len(group) for group in groups]))

        # sum weighted Gini index for each group
        gini = 0.0
        # 2) Summate weighed Gini index for each group.
        for group in groups:

            # 2.1) Calcuate size of a group.
            size = float(len(group))
            # 2.2) Catch to avoid division by zero
            if size == 0:
                continue

            # 2.3) Summate score

            p = [row[-1] for row in group].count(classes[0]) / size 
            gini += ((1.0 - p)*(2*p)) * (size / n_instances)
        return gini

    # CSCE 633 Entropy Index Func
    def getEntropy(self,groups, classes):
        #print("Getting entropy")
        # 1) Get count of all instances in split groups.
        n_instances = 0
        for group in groups:
            n_instances += len(group)

        # Old:
        #n_instances = float(sum([len(group) for group in groups]))

        # sum weighted Gini index for each group
        entropy = 0.0

        # 2) Summate weighed Gini index for each group.
        for group in groups:

            # 2.1) Calcuate size of a group.
            size = float(len(group))
            # 2.2) Catch to avoid division by zero
            if size == 0:
                continue

            # 2.3) Summate score

            p = [row[-1] for row in group].count(classes[0]) / size

            #print(p)

            entropyValGrp = ( -p*np.log(p) ) - ( (1-p) * np.log(1-p) )

            entropy += entropyValGrp * (size / n_instances)
        return entropy

    '''
    Functions to build tree
    
    '''
    '''
    This function determines for a given group
    which outcome is more frequent. Used for a 
    terminal leaf.
    '''
    # Create a terminal node value
    def setTerm(self,group):
        
        outcomes = [row[-1] for row in group]
        return max(set(outcomes), key=outcomes.count)
           
    
    # Create child splits for a node or make terminal
    def split(self,node, max_depth, min_size, depth,in_sc):
        #print("Starting a new split.")
        #1
        #left, right = node['groups']
        #del(node['groups'])
        
        
        left,right = node.bestSplitGroups
        node.bestSplitGroups = None
        
        #2
        if not left or not right:
            #print("This is a leaf node with no split.")
            node.leftBranch = rootObject(999,999,None)
            node.rightBranch = rootObject(999,999,None)
            
            #node.leftBranch = node
            #node.rightBranch = node
            
            node.leftBranch.bestSplitIndex = node.rightBranch.bestSplitIndex = node.bestSplitIndex 
            node.leftBranch.bestSplitVal = node.rightBranch.bestSplitVal = self.setTerm(left + right)
            #print("Left branch best split value: %s" %(node.leftBranch.bestSplitVal))
            #print("Right branch best split value: %s" %(node.rightBranch.bestSplitVal))
            return

        #3
        if depth >= max_depth:
            #print("This is a leaf node with two splits.")
            node.leftBranch = rootObject(999,999,None)
            node.rightBranch = rootObject(999,999,None)
            
            node.leftBranch.bestSplitIndex = node.rightBranch.bestSplitIndex = node.bestSplitIndex 
            node.leftBranch.bestSplitVal, node.rightBranch.bestSplitVal = self.setTerm(left), self.setTerm(right)
            #print("Left branch best split value: %s" %(node.leftBranch.bestSplitVal))
            #print("Right branch best split value: %s" %(node.rightBranch.bestSplitVal))
            return

        #4
        if len(left) <= min_size:
            #print("This is a left branch with a leaf node.")
            node.leftBranch = rootObject(999,999,None)
            node.leftBranch.bestSplitVal = self.setTerm(left)
            #print("Left branch best split value: %s" %(node.leftBranch.bestSplitVal))
        else:
            #print("This is a left branch that will split.")
            node.leftBranch = rootObject(999,999,None)
            b_index, b_value, b_groups = self.getBestSplit(left,node.leftBranch.bestSplitIndex,self.sc)
            node.leftBranch.bestSplitIndex = b_index
            node.leftBranch.bestSplitVal = b_value
            node.leftBranch.bestSplitGroups = b_groups
            self.split(node.leftBranch, max_depth, min_size, depth+1,in_sc)
            #print("Left branch best split index: %s" %(node.leftBranch.bestSplitIndex))
            #print("Left branch best split value: %s" %(node.leftBranch.bestSplitVal))

        #5
        if len(right) <= min_size:
            #print("This is a right branch with a leaf node.")
            node.rightBranch = rootObject(999,999,None)
            node.rightBranch.bestSplitVal = self.setTerm(right)
            #print("Right branch best split value: %s" %(node.rightBranch.bestSplitVal))
            
        else:
            #print("This is a right branch that will split.")
            node.rightBranch = rootObject(999,999,None)
            b_index, b_value, b_groups = self.getBestSplit(right,node.rightBranch.bestSplitIndex,self.sc)
            node.rightBranch.bestSplitIndex = b_index
            node.rightBranch.bestSplitVal = b_value
            node.rightBranch.bestSplitGroups = b_groups
            self.split(node.rightBranch, max_depth, min_size, depth+1,in_sc)
            #print("Right branch best split index: %s" %(node.rightBranch.bestSplitIndex))
            #print("Right branch best split value: %s" %(node.rightBranch.bestSplitVal))

    def printTree(self, node, depth=0):
        if ((node.leftBranch != None) and (node.rightBranch != None)):
            print('%s[X%d < %.3f]' % ((depth*' ', (node.bestSplitIndex+1), node.bestSplitVal)))
            self.print_tree(node.leftBranch, depth+1)
            self.print_tree(node.rightBranch, depth+1)
        else:
            print('%s[%s]' % ((depth*' ', node.bestSplitVal)))        
    
    def getTree(self):
        self.printTree(self.rootNode, 0)
        
    def predict(self,node, row):
        #print("Row index: %s" % (node.bestSplitIndex))
        #print("Row %s" % (row))
        #print("///////////////")
        while(node.bestSplitIndex != 999):
            if row[node.bestSplitIndex] < node.bestSplitVal:
                if (node.leftBranch != None):
                    return self.predict(node.leftBranch, row)
                else:
                    #print("Best split val: %s" % (node.bestSplitVal))
                    return node.bestSplitVal
            else:
                if (node.rightBranch != None):
                    return self.predict(node.rightBranch, row)
                else:
                    #print("Best split val: %s" % (node.bestSplitVal))
                    return node.bestSplitVal
    
    def getPredict(self,test):
        #  predict with a stump
        #stump = {'index': 0, 'right': 1, 'value': 6.642287351, 'left': 0}
        score = 0.0
        iterVal = 0
        for row in test:
            iterVal +=1
            prediction = self.predict(self.rootNode, row)
            #print('Expected=%s, Got=%s' % (row[-1], prediction))
            if(row[-1] == prediction):
                score += 1
        avgScore = ( ( score/float( iterVal ) ) * 100.0 )
        print("Average performance: %s %%" %(avgScore))
        return avgScore
            
```

```python
from random import seed
from random import randrange
from csv import reader
from sklearn.model_selection import train_test_split

# Load a CSV file
def load_csv(filename):
    file = open(filename, "rb")
    lines = reader(file)
    dataset = list(lines)
    return dataset
 
# Convert string column to float
def str_column_to_float(dataset, column):
    for row in dataset:
        row[column] = float(row[column].strip())

filename = '.../hw2_question1.csv'
dataset = load_csv(filename)
# convert string attributes to integers
for i in range(len(dataset[0])):
	str_column_to_float(dataset, i)

########
trainProp = (2/3.0)
testProp = (1/3.0)
together_train,together_test = train_test_split(dataset, test_size = testProp, train_size =  trainProp)
```
