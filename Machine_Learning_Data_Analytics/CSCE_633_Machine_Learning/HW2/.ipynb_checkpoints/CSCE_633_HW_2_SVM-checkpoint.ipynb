```python
def SVM_CV(x_dataset,y_dataset,splits,svclassifier):
    accuracy_score = 0.0
    overallTrainTime = 0.0
    kf = KFold(splits) 
    KFold(splits, random_state=None, shuffle=False)
    i = 1
    
    
    for train_index, test_index in kf.split(x_dataset):
        #print("TRAIN:", train_index, "TEST:", test_index)
        X_train, X_test = x_dataset[train_index], x_dataset[test_index]
        y_train, y_test = y_dataset[train_index], y_dataset[test_index]
        t0 = time()
        #print(X_train)
        #print(y_train)
        svclassifier.fit(X_train, y_train)
        trainTimeForCurrentFold = time() - t0
        #print("Training time: %0.3fs" % (trainTimeForCurrentFold))
        
        #Making predictions
        y_pred = svclassifier.predict(X_test)
        #print(confusion_matrix(y_test,y_pred))  
        #print(classification_report(y_test,y_pred))
        #f1 = sklearn.metrics.f1_score(y_test,y_pred)
        acc = sklearn.metrics.accuracy_score(y_test,y_pred)
        #print("F1 score for current fold %s : %s" %(i,f1))
        i += 1
        accuracy_score += acc
        overallTrainTime += trainTimeForCurrentFold
    avg_acc = accuracy_score/float(splits)
    avgTime = overallTrainTime/float(splits)
    print("Average F1 score across %s folds: %s" % (splits,avg_acc))
    print("Average train time across %s folds: %0.3f" % (splits,avgTime))
    return avg_acc,avgTime

```

```python
'''
Use linear SVM in LIBSVM:
'''





'''
i_iter = 0
for train, test in kfold.split(data):
    i_iter += 1
    #print("Iteration: %s" % (i_iter))
    #print('train: %s, test: %s' % (data[train], data[test]))
    x = data[train][:,0:30]
    y = data[train][:,30]
    #print(data[train])
    #print(data[train][:,30])
    svclassifier = SVC(C=25,kernel='linear') 
    SVC_CV(x,y,svclassifier)
'''    

    
def cvForSVM_kFold(data,svclassifier,verboseVal):
    kfold = KFold(3, True, 1)
    i_iter = 0
    avgTrainTime = 0.0
    avgAcc = 0.0
    for train, test in kfold.split(data):
        i_iter += 1
        #print("Iteration: %s" % (i_iter))
        #print('train: %s, test: %s' % (data[train], data[test]))
        x = data[train][:,0:30]
        y = data[train][:,30]
        #print(data[train])
        #print(data[train][:,30])
        #svclassifier = SVC(C=25,kernel='linear') 
        trainTime,acc = SVC_CV(x,y,svclassifier)
        if(verboseVal == True):
            print("Train time for fold %s : %s" % (i_iter, trainTime))
            print("Accuracy for fold %s : %s" % (i_iter, acc))
        avgTrainTime += trainTime
        avgAcc += acc
    avgTrainTime = avgTrainTime/3.0
    avgAcc =  ( avgAcc/3.0 ) * (100)
    print("$$$$$$$$")
    print("Average train time for current classifer: %s" % (avgTrainTime))
    print("Average accuracy for current classifer: %s %%" % (avgAcc))
    print("$$$$$$$$")
    return avgTrainTime,avgAcc


# C value set at 1
print("C value: 1")
svclassifier = SVC(C=1,kernel='linear') 
cTime_1,cAcc_1 = cvForSVM_kFold(data,svclassifier,False)

# C value set at 5
print("C value: 5")
svclassifier = SVC(C=5,kernel='linear') 
cTime_5,cAcc_5 = cvForSVM_kFold(data,svclassifier,False)

# C value set at 10
print("C value: 10")
svclassifier = SVC(C=10,kernel='linear') 
cTime_10,cAcc_10 =cvForSVM_kFold(data,svclassifier,False) 

# C value set at 15
print("C value: 15")
svclassifier = SVC(C=15,kernel='linear') 
cTime_15,cAcc_15 = cvForSVM_kFold(data,svclassifier,False)

# C value set at 20
print("C value: 20")
svclassifier = SVC(C=20,kernel='linear') 
cTime_20,cAcc_20 = cvForSVM_kFold(data,svclassifier,False)

bestCVal = 0
bestCVal_acc = max([cAcc_1,cAcc_5,cAcc_10,cAcc_15,cAcc_20])
if(bestCVal_acc == cAcc_1):
    bestCVal = 1
elif(bestCVal_acc == cAcc_5):
    bestCVal = 5
elif(bestCVal_acc == cAcc_10):
    bestCVal = 10
elif(bestCVal_acc == cAcc_15):
    bestCVal = 15
elif(bestCVal_acc == cAcc_20):
    bestCVal = 20
print("Best C value for accuracy: %s" % (bestCVal))




# TEST SET (2/3)
print("$$$$$$$$$$$$$")
print("Using best C value on test set:")
svclassifier = SVC(C=bestCVal,kernel='linear')
dtest = together_test.values
x = dtest[:,0:30]
y = dtest[:,30]


test_trainTime,test_acc = SVC_CV(x,y,svclassifier)
test_acc = test_acc *(100.0)
print("Train time for C value %s : %s" % (bestCVal, test_trainTime))
print("Accuracy for C value %s : %s %%" % (bestCVal, test_acc))
```
