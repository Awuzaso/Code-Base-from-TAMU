```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# define sinusoid of length T (sec), with fundamental frequency F0 (Hz), sampled at Fs (Hz)
def sinusoid(Fs,F0,T):
    t=np.linspace(0,T,T*Fs)
    return (t,np.sin(2*np.pi*F0*t))



(T,Fs)=(20,10)
(F1,F2)=(1,0.2)
(A,B)=(1,2)
(t,y1)=sinusoid(Fs,F1,T)
(t,y2)=sinusoid(Fs,F2,T)
ys=A*y1+B*y2
plt.figure(1);plt.plot(t,A*y1,'r--')
plt.figure(2);plt.plot(t,B*y2,'bs')
plt.figure(3);plt.plot(t,ys,'g--^')
plt.show()
```

```python
# b) Con

N=100 # FFT length
fft_y1=np.fft.rfft(y1,N)
fft_y2=np.fft.rfft(y2,N)
fft_ys=np.fft.rfft(ys,N)
f = np.linspace(0,(Fs/2),N/2+1)
plt.figure(4);plt.plot(f,np.abs(fft_y1),'r--')
plt.figure(5);plt.plot(f,np.abs(fft_y2),'b--')
plt.figure(6);plt.plot(f,np.abs(fft_ys),'g--')
```

```python
s=0.8 #amplitude of Gaussian noise
ys_n=ys+s*np.random.rand(len(t))
fft_ys_n=np.fft.rfft(ys_n,N)
plt.figure(7);plt.plot(t,ys_n,'r--',t,ys,'g-')
plt.figure(8);plt.plot(f,np.abs(fft_ys_n),'r--',f,np.abs(fft_ys),'g-')
```

```python
def lowpass_filter(x, lowcut, Fs, filter_order):
    nyquist_freq = 0.5 * Fs
    low = lowcut / nyquist_freq
    b, a = signal.butter(filter_order,low, btype="lowpass")
    y = signal.lfilter(b, a, x)
    return (y,b,a)

(ys_n_smooth,b,a) = lowpass_filter(ys_n,1,Fs,1)
w, h = signal.freqs(b, a);
plt.figure(9);plt.plot(w, 20 * np.log10(abs(h)))
plt.figure(10);plt.plot(t,ys_n_smooth,'r--',t,ys_n,'b--',t,ys,'g-')
plt.figure(11);plt.plot(f,np.abs(np.fft.rfft(ys_n_smooth,N)),'r--',f,np.abs(np.fft.rfft(ys_n,N)),'b--',f,np.abs(np.fft.rfft(ys,N)),'g-')
```
